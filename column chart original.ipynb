```python
import pandas as pd

data = pd.read_csv('train.csv')


data.head()
```

```python
d = {'male' : 'M', 'female' : 'F'}
data['Sex'] = data['Sex'].apply(lambda x:d[x])

data['Sex'].head()
```

```python
pclass_gender_survival_count_df = data.groupby(['Pclass', 'Sex'])['Survived'].sum()
pclass_gender_survival_count_df
```

```python
import matplotlib.pyplot as plt 

fig = plt.figure(figsize=(30,10))


plt.subplot2grid((2,3),(0,0))
data.Survived.value_counts().plot(kind='bar', alpha=0.5)
plt.title('Sobrevivieron - cuenta total -')

plt.subplot2grid((2,3),(0,1))
data.Survived.value_counts(normalize = True).plot(kind='bar', alpha=0.5)
plt.title('Sobrevivieron - porcentaje total -')





plt.show()
```

```python


fig = plt.figure(figsize=(30,10))

data.Sex[data.Survived == 1].value_counts(normalize = True).plot(kind='barh', alpha=0.5) ###aqui no me permitia agregar un color para identificar
plt.title('Sobrevivieron - Mujer - Hombre -')

plt.show()
```

```python
import matplotlib.pyplot as plt 

fig = plt.figure(figsize=(10,5))

data.Pclass[data.Survived == 1].value_counts(normalize = True).plot(kind='bar', alpha=0.5) ###aqui no me permitia agregar un color para identificar
plt.title('Sobrevivientes por la clase de ticket')

plt.show()
```
