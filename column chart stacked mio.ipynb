```python
# aqui lo que hacemos es motrar especies junto con el sexo ordenado por hindfoot_length by_site_sex = surveys_df.groupby(['species_id','sex'])
site_sex_count = by_site_sex['hindfoot_length'].sum()
spc = site_sex_count.unstack()
s_plot = spc.plot(kind='bar',stacked=True,title="ejemplo 2")
s_plot.set_ylabel("genus")
s_plot.set_xlabel("species_id")
```
